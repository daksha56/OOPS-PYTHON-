##📌 TOPIC 1: Classes & Objects
📖 Explanation



A class is a blueprint. An object is the real thing built from that blueprint. Every object gets its own copy of the data.

```python
class Dog:
    def __init__(self, name, breed):
        self.name = name
        self.breed = breed

    def bark(self):
        return f"{self.name} says: Woof!"

dog1 = Dog("Bruno", "Labrador")
dog2 = Dog("Max", "Poodle")

print(dog1.bark())  # Bruno says: Woof!
print(dog2.bark())  # Max says: Woof!
```

#📌 TOPIC 2: self & Instance Properties
📖 Explanation
self refers to the current object. It lets each object store and access its own data separately from other objects.

```python
class Person:
    def __init__(self, name, age):
        self.name = name   # belongs to THIS object
        self.age = age

    def greet(self):
        return f"I am {self.name}, aged {self.age}"

p1 = Person("Daksh", 20)
p2 = Person("Rahul", 25)

print(p1.greet())  # I am Daksh, aged 20
print(p2.greet())  # I am Rahul, aged 25
```

#📌 TOPIC 3: Class Properties & @classmethod
📖 Explanation

Instance property → belongs to ONE object (self.x)
Class property → shared across ALL objects (ClassName.x)
@classmethod → a method that works on the class itself, not an instance. Uses cls instead of self

```python
class Employee:
    company = "TechCorp"      # class property (shared)
    employee_count = 0

    def __init__(self, name, salary):
        self.name = name          # instance property
        self.salary = salary
        Employee.employee_count += 1

    @classmethod
    def get_company(cls):
        return f"Company: {cls.company}, Total Employees: {cls.employee_count}"

e1 = Employee("Daksh", 50000)
e2 = Employee("Rahul", 60000)

print(e1.name)                  # Daksh (instance)
print(Employee.company)         # TechCorp (class)
print(Employee.get_company())   # Company: TechCorp, Total Employees: 2
```

#📌 TOPIC 4: Dunder/Magic Methods
📖 Explanation
Dunder = Double underscore. These are special built-in methods Python calls automatically in certain situations.

```python
class Bag:
    def __init__(self, items):
        self.items = items

    def __str__(self):
        return f"Bag contains: {self.items}"

    def __len__(self):
        return len(self.items)

    def __add__(self, other):
        return Bag(self.items + other.items)

    def __eq__(self, other):
        return len(self.items) == len(other.items)

b1 = Bag(["book", "pen"])
b2 = Bag(["water", "snack", "phone"])

print(b1)           # Bag contains: ['book', 'pen']
print(len(b1))      # 2
b3 = b1 + b2
print(b3)           # Bag contains: ['book', 'pen', 'water', 'snack', 'phone']
print(b1 == b2)     # False
```

#📌 TOPIC 5: @staticmethod
📖 Explanation
A static method doesn't use self or cls. It's just a regular function that lives inside a class for organizational purposes. It doesn't access or modify class/instance data.

```python
class MathUtils:
    @staticmethod
    def add(a, b):
        return a + b

    @staticmethod
    def is_even(n):
        return n % 2 == 0

print(MathUtils.add(5, 3))      # 8
print(MathUtils.is_even(4))     # True
```

#📌 TOPIC 6: Encapsulation
📖 Explanation
Encapsulation = hiding internal data from outside access. We control what can be read or changed using:

```python
class BankAccount:
    def __init__(self, owner, balance):
        self.owner = owner
        self.__balance = balance    # private

    def get_balance(self):          # getter
        return self.__balance

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount

    def withdraw(self, amount):
        if amount <= self.__balance:
            self.__balance -= amount
        else:
            print("Insufficient funds!")

acc = BankAccount("Daksh", 10000)
acc.deposit(5000)
acc.withdraw(3000)
print(acc.get_balance())    # 12000
#print(acc.__balance)        # ❌ AttributeError - private!
```

#📌 TOPIC 7: Inheritance
📖 Explanation
Inheritance lets a child class reuse code from a parent class. The child gets all parent methods and can add its own or override them.

```python
class Animal:
    def __init__(self, name, sound):
        self.name = name
        self.sound = sound

    def speak(self):
        return f"{self.name} says {self.sound}"

class Dog(Animal):
    def fetch(self):
        return f"{self.name} fetches the ball!"

class Cat(Animal):
    def purr(self):
        return f"{self.name} purrs..."

dog = Dog("Bruno", "Woof")
cat = Cat("Kitty", "Meow")

print(dog.speak())   # Bruno says Woof  ← inherited
print(dog.fetch())   # Bruno fetches the ball!
print(cat.speak())   # Kitty says Meow  ← inherited
print(cat.purr())    # Kitty purrs...
```

#📌 TOPIC 8: super()
📖 Explanation
super() lets a child class call the parent's __init__ (or any method) without repeating code.

```python
class Animal:
    def __init__(self, name):
        self.name = name

class Dog(Animal):
    def __init__(self, name, breed):
        super().__init__(name)    # calls Animal's __init__
        self.breed = breed

    def info(self):
        return f"{self.name} is a {self.breed}"

d = Dog("Bruno", "Labrador")
print(d.info())   # Bruno is a Labrador
```

#📌 TOPIC 9: Polymorphism
📖 Explanation
Polymorphism = same method name, different behavior depending on the object. One interface, many forms.

```python
class Shape:
    def area(self):
        pass

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius
    def area(self):
        return 3.14 * self.radius ** 2

class Rectangle(Shape):
    def __init__(self, length, width):
        self.length = length
        self.width = width
    def area(self):
        return self.length * self.width

shapes = [Circle(5), Rectangle(4, 6)]

for shape in shapes:
    print(shape.area())   # Each calls its OWN area()
# 78.5
# 24
```

#📌 TOPIC 10: Abstraction
📖 Explanation
Abstraction = hiding implementation details, showing only what's necessary. We use ABC (Abstract Base Class) to force child classes to implement certain methods.

```python
from abc import ABC, abstractmethod

class Payment(ABC):
    @abstractmethod
    def pay(self, amount):   # must be implemented by children
        pass

class CreditCard(Payment):
    def pay(self, amount):
        return f"Paid ₹{amount} via Credit Card"

class UPI(Payment):
    def pay(self, amount):
        return f"Paid ₹{amount} via UPI"

# Payment()  ← ❌ Can't instantiate abstract class

c = CreditCard()
u = UPI()
print(c.pay(500))   # Paid ₹500 via Credit Card
print(u.pay(300))   # Paid ₹300 via UPI
```
