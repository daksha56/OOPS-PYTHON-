#📌 TOPIC 1: Classes & Objects
Question:
Create a class `Laptop` with:
- Properties: brand, ram, storage
- Method specs() → returns:
  "Brand: {brand}, RAM: {ram}GB, Storage: {storage}GB"
- Method is_powerful() → returns True if ram >= 16

Create 2 laptop objects and test both methods.

```python
class laptop():
  def __init__(self,brand,ram,storage):
    self.brand=brand
    self.ram=ram
    self.storage=storage
  def specs(self):
    return f"Brand: {self.brand}, RAM: {self.ram}GB, Storage: {self.storage}GB"
  def is_powerful(self):
    if self.ram>=16:
      return True
    else:
        return False
lap1=laptop("HP",18,500)
lap2=laptop("DEll",12,320)
print(lap1.specs())
print(lap2.specs())
print(lap2.is_powerful())
print(lap1.is_powerful())
```

#📌 TOPIC 4: Dunder/Magic Methods
📝 Your Question
Create a class `Cart` with:
- Property: items (list of products)
- __str__  → "Cart has: {items}"
- __len__  → number of items
- __add__  → merge 2 carts into one
- __eq__   → True if both carts have same number of items

Create 2 carts, merge them, check equality.

```python
class Cart():
  def __init__(self,items):
    self.items=items
  def __str__(self):
    return f"Cart has: {self.items}"
  def __len__(self):
    return len(self.items)
  def __add__(self,other):
    return Cart(self.items+other.items)
  def __eq__(self,other):
    return len(self.items)==len(other.items)
c1=Cart(["bag","shampoo","fruits"])
c2=Cart(["biscuits","soap","vegetables"])
print(c1)
print(c1==c2)
print(c2)
print(len(c1))
print(len(c2))
print(c1+c2)
```

#📌 TOPIC 9: Polymorphism
Create a base class `Discount` with method calculate(price)

Create 3 child classes:
- `FlatDiscount` → deducts flat 200 from price
- `PercentDiscount` → deducts 10% from price
- `NoDiscount` → returns original price

Store all 3 in a list and loop through,
calling calculate(1000) on each.

```python
class Discount:
  def calculate(self, price):
    raise NotImplementedError("Subclasses must implement this method")

class FlatDiscount(Discount):
  def calculate(self, price):
    return price - 200

class PercentDiscount(Discount):
  def calculate(self, price):
    return price * 0.90  # 10% deduction

class NoDiscount(Discount):
  def calculate(self, price):
    return price

# Create instances of the discount classes
flat = FlatDiscount()
percent = PercentDiscount()
no_discount = NoDiscount()

# Store them in a list
discounts = [flat, percent, no_discount]

# Loop through and call calculate(1000) on each
for d in discounts:
  print(f"Discount type: {type(d).__name__}, Calculated price: {d.calculate(1000)}")
```

#📌 TOPIC 10: Abstraction
Create an abstract class `Appliance` with:
- @abstractmethod turn_on()
- @abstractmethod turn_off()

Create 2 child classes:
- `WashingMachine` → turn_on() prints "Washing started"
                  → turn_off() prints "Washing done"
- `AirConditioner` → turn_on() prints "Cooling started at 24°C"
                   → turn_off() prints "AC turned off"

Create objects and call both methods on each.

```python
from abc import ABC,abstractmethod
class appliance(ABC):
  @abstractmethod
  def turn_on(self):
    pass
  @abstractmethod
  def turn_off(self):
    pass
class washing_machine(appliance):
  def turn_on(self):
    print("Washing Started")
  def turn_off(self):
    print("washing done")
class air_conditioner(appliance):
  def turn_on(self):
    print("Cooling started at 24 c")
  def turn_off(self):
    print("AC turned off")
wash1=washing_machine()
wash1.turn_on()
wash1.turn_off()
ac=air_conditioner()
ac.turn_on()
ac.turn_off()
```
