##📌 TOPIC 1: Classes & Objects
📖 Explanation



A class is a blueprint. An object is the real thing built from that blueprint. Every object gets its own copy of the data.

```python
class Dog:
    def __init__(self, name, breed):
        self.name = name
        self.breed = breed

    def bark(self):
        return f"{self.name} says: Woof!"

dog1 = Dog("Bruno", "Labrador")
dog2 = Dog("Max", "Poodle")

print(dog1.bark())  # Bruno says: Woof!
print(dog2.bark())  # Max says: Woof!
```
